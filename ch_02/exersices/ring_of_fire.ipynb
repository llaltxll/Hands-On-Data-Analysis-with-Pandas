```python
import pandas as pd
```

```python
df_quakes = pd.read_csv("../data/parsed.csv")
```

```python
sr_ring_places = pd.Series(name="place",data=[
   "Bolivia",
   "Chile",
   "Ecuador",
   "Peru",
   "Costa Rica",
   "Guatemala",
   "Mexico",
   "Japan",
   "Philippines",
   "Indonesia",
   "New Zealand",
   "Antarctic",
   "Canada",
   "Fiji",
   "Alaska",
   "Washington",
   "California",
   "Russia",
   "Taiwan",
   "Tonga",
   "Kermadec Islands"])
```

```python
df_ring = df_quakes[df_quakes.parsed_place.isin(sr_ring_places)]
df_ring.parsed_place.value_counts()
sr_not_found = sr_ring_places[~sr_ring_places.isin(df_ring.parsed_place)]
sr_not_found
```

```python
df_antarctic = df_quakes[df_quakes.parsed_place.str.contains("Antarctic")]
df_antarctic
```

```python
df_ring = pd.concat([df_ring, df_antarctic])
df_ring
```
