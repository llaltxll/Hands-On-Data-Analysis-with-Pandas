#### Find the 95th percentile of earthquake magnitude in Japan using the magType of 'mb'.

```python
import pandas as pd
```

```python
df_quakes = pd.read_csv("../data/parsed.csv")
df_quakes.info()
```

```python
df_quakes.mag.describe()
```

```python
df_quakes.head()
```

```python
df_quakes.parsed_place.unique()
```

```python
df_japan_mag = df_quakes[(df_quakes.parsed_place == "Japan") & \
                                    (df_quakes.magType == "mb")]\
                                    [["parsed_place", "mag", "magType"]]
df_japan_mag.describe()
```

```python
df_japan_mag.mag.quantile(q=.95)
```
